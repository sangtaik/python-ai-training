```python
import cv2
import numpy as np
import sys

# 4.5.5 버전이 기준이다.
cv2.__version__
```

```python
# 이미지를 창에 띄우는 코드
img = cv2.imread('./fig/puppy.bmp', flags = cv2.IMREAD_COLOR)

if img is None: # 이미지가 없으면 프로그램 종료 
    print('image read failed')
    sys.exit()
    
img_resize = cv2.resize(img, (300, 200), interpolation = cv2.INTER_AREA)
cv2.imwrite('./fig/image_resize.png', img_resize)
cv2.namedWindow('image', cv2.WINDOW_NORMAL) # AUTOSIZE랑 다르게 창을 조절할 수 있다.
cv2.imshow('image', img)
cv2.imshow('img_resize', img_resize)
while True:
    keys = cv2.waitKey()
    if keys == 27 or keys == ord('q'):
        break
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt

imgBGR = cv2.imread('./fig/puppy.bmp', cv2.IMREAD_COLOR)
imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)
imgGRAY = cv2.cvtColor(imgBGR, cv2.IMREAD_GRAYSCALE)
if imgBGR is None:
    print('image read failed')
    sys.exit()
    
fig, axes = plt.subplots(1, 3, figsize = (8, 8))
axes[0].imshow(imgBGR), axes[0].axis('off')
axes[1].imshow(imgRGB), axes[1].axis('off')
axes[2].imshow(imgGRAY, cmap = 'gray'), axes[2].axis('off')

plt.show()

while True:
    if cv2.waitKey() == 27:
        break
        
cv2.destoryAllWindows()
```

```python
# 색을 기준으로 마스크(mask)를 생성
src = cv2.imread('./fig/palette.png')
src_dark = cv2.imread('./fig/palette_dark.png')

# 색을 기준으로 뽑아내는 함수
# inRange(src, lowerb, upperb[, dst]) -> dst
# src : 입력 영상
# lowerb : 하한값 (0, 0, 0)
# upperb : 상한값 (255, 255, 255) 
# (0, 150, 0), (100, 255, 100) 이면 0 ~ 100, 140 ~ 255, 0 ~ 100 범위로 (BGR이든 RGB든 뽑힌다.
dst_bgr = cv2.inRange(src, (0, 150, 0), (100, 255, 100))
dst_bgr_dark = cv2.inRange(src_dark, (0, 150, 0), (50, 255, 50))
print(dst_bgr.shape)
print(dst_bgr_dark.shape)

cv2.imshow('src', src)
cv2.imshow('dst_bgr', dst_bgr)
cv2.imshow('dst_bgr_dark', dst_bgr_dark)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 크로마키 사용하여 합성하기

# 트랙바를 이용하여 색 뽑아보기
# 색을 기준으로 마스크(mask)를 생성   HSV를 사용하여

    
src = cv2.imread('./fig/autumn.jpg')
dst = cv2.imread('./fig/chromakey.png')

h, w = dst.shape[:2]
src = cv2.resize(src, (w, h), cv2.INTER_AREA)

dst_hsv = cv2.cvtColor(dst, cv2.COLOR_BGR2HSV)

# 마스크 추출
mask = cv2.inRange(dst_hsv, (50, 50, 0), (80, 255, 255))

cv2.copyTo(src, mask, dst)

cv2.imshow('src', src)
cv2.imshow('dst', dst)
cv2.imshow('mask', mask)


cv2.waitKey()
cv2.destroyAllWindows()
```

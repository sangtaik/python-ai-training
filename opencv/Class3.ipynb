```python
img_list = os.listdir('./fig/images/')

img_lists = []

for i in img_list:
    file_path = './fig/images/' + i
    img_lists.append(file_path)
    
img_lists

cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.setWindowProperty('image', cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)

idx = 0

while True:
    img = cv2.imread(img_lists[idx])
    
    cv2.imshow('image', img)
    
    if cv2.waitKey(3000) == 27:
        break
        
    idx += 1
    
    if idx >= len(img_lists):
        idx = 0
        
cv2.destroyAllWindows()
```

```python
# 큰 초원에 송아지를 옮기기
import numpy as np
import cv2
import sys 

src = cv2.imread('./fig/cow.png')
dst = cv2.imread('./fig/green.png')
src_gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)

h, w = src.shape[:2] # 소 그림의 가로, 세로 크기를 가져온다.
print(h, w)

# 마스크 만들기 : 트레스 홀드
_, mask = cv2.threshold(src_gray, 235, 255, cv2.THRESH_BINARY_INV)

# 배경 화면에서 소의 그림만큼 위치를 따온다.
dst2 = dst[1000-h:1000, 1400-w:1400]

# 그 위치에 소를 집어 넣는다.
cv2.copyTo(src, mask, dst2)

if src is None or dst is None:
    print('image read failed')
    sys.exit()
    
cv2.imshow('src', src)
cv2.imshow('dst', dst)
#cv2.imshow('dst2', dst2)


cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 알파 채널 사용하기
import numpy as np
import cv2
import sys 

src = cv2.imread('./fig/puppy.bmp')
# ALPHA 채널 사용 시 , IMREAD_UNCHANGED 사용
sunglass_4 = cv2.imread('./fig/imgbin_sunglasses_1.png', cv2.IMREAD_UNCHANGED)

if src is None or sunglass is None:
    print('image read failed')
    sys.exit()

# 썬글라스 크기 조절
sunglass_4 = cv2.resize(sunglass_4, (300, 150))

sunglass = sunglass_4[:,:, 0:3]
mask = sunglass_4[:,:,-1] # 알파채널 
print(src.shape)
print(sunglass.shape)

# h, w 계산
h, w = mask.shape[:2]
crop = src[120:120+h, 220:220+w]

# 그 위치에 안경을 집어 넣는다.
cv2.copyTo(sunglass, mask, crop)

crop[mask > 0]  = (0, 0, 255)    
    
    
cv2.imshow('src', src)
cv2.imshow('mask', mask)
cv2.imshow('sunglass', sunglass)
cv2.imshow('crop', crop)
cv2.waitKey()
cv2.destroyAllWindows()
```

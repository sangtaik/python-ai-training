```python
import numpy as np
import cv2
import sys 

import matplotlib.pyplot as plt
import os
```

```python
## keyboard event and redo by stack

img = cv2.imread('./fig/puppy.bmp', cv2.IMREAD_GRAYSCALE)

if img is None:
    print('image read failed')
    sys.exit()

cv2.imshow('img', img)

stack = []
indexes = []

# 원본 이미지 저장하기
stack.append(img)
indexes.append('origin')
while True:
    key = cv2.waitKey()
    if key == 27:
        break
    elif key == ord('i'): # 반전 효과
        img = 255 -img
        stack.append(img)
        indexes.append('i')
        cv2.imshow('img', img)
    elif key == ord('e'):
        img = cv2.Canny(img, 50, 150)
        stack.append(img)
        indexes.append('e')
        cv2.imshow('img', img)
    elif key == ord('r'):
        if len(stack) > 0:
            stack.pop()
            indexes.pop()
            img = stack[len(stack) -1]
        cv2.imshow('img', img)
    elif key == ord('s'):
        cv2.imwrite('./fig/puppy_output.png', img)
    print(len(stack))
    print(indexes)
        
cv2.destroyAllWindows()
```

```python
## keyboard event 방법2:

img = cv2.imread('./fig/puppy.bmp', cv2.IMREAD_GRAYSCALE)

if img is None:
    print('image read failed')
    sys.exit()

cv2.imshow('img', img)

while True:
    key = cv2.waitKey()
    img2 = img.copy()
    if key == 27:
        break
    elif key == ord('i'): # 반전 효과
        img = 255 -img
        cv2.imshow('img', img)
    elif key == ord('e'):
        img = cv2.Canny(img, 50, 150)
        cv2.imshow('img', img)
    elif key == ord('r'):
        img = img2.copy()
        cv2.imshow('img', img)
    elif key == ord('s'):
        cv2.imwrite('./fig/puppy_output.png', img)
        
cv2.destroyAllWindows()
```

```python
# mouse call

# on

def mouse_call(event, x, y, flags, param):
    
    global oldx, oldy
    
    if event == cv2.EVENT_LBUTTONDOWN:
        oldx, oldy = x, y
        #print('left button down = {}, {}'.format(x, y))
        
    elif event == cv2.EVENT_MOUSEMOVE:
        if flags & cv2.EVENT_FLAG_LBUTTON: # 버튼을 누르면 동작. &는 ==랑 같은 의미로 쓰임
            #print('Mouse moved = {}, {}'.format(x, y))
            cv2.line(img, (oldx, oldy), (x, y), (0, 0, 0), 4, cv2.LINE_AA)
            cv2.imshow('img', img)
            oldx, oldy = x, y
        
    

img = np.ones((600, 800, 3), np.uint8)*255

cv2.namedWindow('img')
# setMouseCallback(windowName, onMouse [, param]) -> None
    # windowName = 마우스 이벤트를 수행할 창 이름
    # onMouse : callback 함수 이름
    # param : 콜백 함수에 전달할 데이터
cv2.setMouseCallback('img', mouse_call, img)

while True:
    key = cv2.waitKey()
    if key == 27:
        break
    elif key == ord('s'):
        cv2.imwrite('./fig/sign_output.png', img)

cv2.destroyAllWindows()
```

```python
# 웹캠 연동하기
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print('Videocap open failed')
    sys.exit()
    
while True:
    ret, frame = cap.read()
    
    if not ret:
        print('Video read failed')
        break
        
    cv2.imshow('frame', frame)
    
    if cv2.waitKey(30) == 27:
        break
        
# 캠을 반드시 닫아준다. 계속 프로그램이 점유해서는 안되기 때문에
cap.release()
cv2.destroyAllWindows()
```

```python
# 웹캠 연동하기
# 엣지 보여주기
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print('Videocap open failed')
    sys.exit()
    
while True:
    ret, frame = cap.read()
    
    if not ret:
        print('Video read failed')
        break
        
    edge = cv2.Canny(frame, 50, 150)
    
    cv2.imshow('frame', frame)
    cv2.imshow('edge', edge)
    
    if cv2.waitKey(30) == 27:
        break
        
# 캠을 반드시 닫아준다. 계속 프로그램이 점유해서는 안되기 때문에
cap.release()
cv2.destroyAllWindows()
```

```python
# 영상 저장하기

cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print('Video open failed')
    sys.exit()
    
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS)*0.7)

# 동영상 저장 코덱을 지정해준다.
# 'D', 'i' 이런 식으로 글자를 풀어줌
fourcc = cv2.VideoWriter_fourcc(*'DIVX') 

print(w, h , fps)

out = cv2.VideoWriter('output.avi', fourcc ,fps, (w, h))
outedge = cv2.VideoWriter('outputedge.avi', fourcc ,fps, (w, h))


while True:
    
    ret, frame = cap.read()
    
    if not ret :
        print('Video read failed')
        break
    edge = cv2.Canny(frame, 50, 150)
    edge = cv2.cvtColor(edge, cv2.COLOR_GRAY2BGR) # GRAY 형식을 COLOR로 변경해줘야 한다.
    out.write(frame)
    outedge.write(edge)
    
    cv2.imshow('image', frame)
    cv2.imshow('edge', edge)
    if cv2.waitKey(120) == 27:
        break
        
cap.release()
out.release()
outedge.release()
cv2.destroyAllWindows()
```

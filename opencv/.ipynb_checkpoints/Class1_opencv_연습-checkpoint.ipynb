```python
import cv2
import numpy as np
import sys

# 4.5.5 버전이 기준이다.
cv2.__version__
```

```python
# 이미지를 창에 띄우는 코드
img = cv2.imread('./fig/puppy.bmp', flags = cv2.IMREAD_COLOR)

if img is None: # 이미지가 없으면 프로그램 종료 
    print('image read failed')
    sys.exit()


print(type(img))

# 640 x 480 (width, height) 그림이지만 numpy이기 때문에 480, 640로 들어온다.
# 좌표계가 다르기 때문이다.
print(img.shape)
print(img.dtype)
cv2.namedWindow('image_color')
cv2.imshow('image_color', img)

# 키를 입력받는 것을 기다린다. 키보드 키를 입력 받으면 창이 닫힌다.
# 창이 닫히는데는 다소 시간이 걸린다.
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img = cv2.imread('./fig/PD_MRI.jpeg', flags = cv2.IMREAD_GRAYSCALE)

if img is None: # 이미지가 없으면 프로그램 종료 
    print('image read failed')
    sys.exit()

#
reduced_mri = cv2.resize(img, (300, 300), interpolation = cv2.INTER_AREA)
# 다른이름으로 저장하기    
cv2.imwrite('./fig/PD_MRI_resize.jpeg', reduced_mri)
    
cv2.namedWindow('image', cv2.WINDOW_NORMAL) # AUTOSIZE랑 다르게 창을 조절할 수 있다.
cv2.imshow('image', img)
cv2.namedWindow('image_resize', cv2.WINDOW_AUTOSIZE)
cv2.imshow('image_resize', reduced_mri)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 실습 2
img = cv2.imread('./fig/puppy.bmp', cv2.IMREAD_COLOR)

if img is None:
    print('Go home')
    sys.exit()
    
cv2.namedWindow('image', cv2.WINDOW_AUTOSIZE)
cv2.imshow('image', img)

cv2.resize(img, (100, 100), interpolation = cv2.INTER_AREA)
cv2.imwrite('./fig/puppy_resize.png', img)

# waitKey는 ms 단위로 창이 꺼지는 타임을 지정할 수 있다.
cv2.waitKey(3000)
cv2.destroyAllWindows()
```

```python
# 실습 3
img = cv2.imread('./fig/puppy.bmp', cv2.IMREAD_COLOR)

if img is None:
    print('Go home')
    sys.exit()
    
cv2.namedWindow('image', cv2.WINDOW_AUTOSIZE)
cv2.imshow('image', img)

# waitKey는 ms 단위로 창이 꺼지는 타임을 지정할 수 있다.
while True:
    keys = cv2.waitKey()
    if keys ==  27 or keys == 13: # ESC키 또는 Enter key를 누르면 창이 꺼진다.
    # if keys == ord('q')
        break

cv2.destroyAllWindows()
```

```python
# matplotlib을 이용한 이미지 출력
import matplotlib.pyplot as plt


imgBGR = cv2.imread('./fig/puppy.bmp', cv2.IMREAD_COLOR)

if imgBGR is None:
    print('image read failed')
    sys.exit()

# OPENCV는 bgr 순서로 읽기 때문에 rgb로 변경해야 한다.
imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)
plt.figure(figsize = (8, 8))
plt.imshow(imgRGB)
plt.axis('off')
plt.show()
# cv2.imshow('image', img)

# while True:
#     key = cv2.waitKey()
#     if key == 27 or key == 113:
#         break
        
# cv2.destroyAllWindows()
```

```python
# matplotlib을 이용한 이미지 출력
import matplotlib.pyplot as plt


imgBGR = cv2.imread('./fig/puppy.bmp', cv2.IMREAD_COLOR)
imgGRAY = cv2.imread('./fig/puppy.bmp', cv2.IMREAD_GRAYSCALE)

if imgBGR is None:
    print('image read failed')
    sys.exit()

# OPENCV는 bgr 순서로 읽기 때문에 rgb로 변경해야 한다.
imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)

fig, axes = plt.subplots(1, 3, figsize = (12, 12))
axes[0].imshow(imgRGB), axes[0].axis('off')
axes[1].imshow(imgBGR), axes[1].axis('off')
axes[2].imshow(imgGRAY, cmap = 'gray'), axes[2].axis('off')
plt.show()
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np


# print(tf.__version__)
mnist = tf.keras.datasets.fashion_mnist
(training_images, training_labels), (test_images, test_labels) = mnist.load_data()

np.set_printoptions(linewidth=200)

# 20   
figsize = [8, 8]     # figure size, inches
ncols = 5
nrows = 5

# create figure (fig), and array of axes (ax)
fig, ax = plt.subplots(nrows=nrows, ncols=ncols, figsize=figsize)

# plot simple raster image on each sub-plot
for i, axi in enumerate(ax.flat):
    # i runs from 0 to (nrows*ncols-1)
    # axi is equivalent with ax[rowid][colid]
    img = training_images[i]
    axi.imshow(img, alpha=0.9)
    # get indices of row/column
    rowid = i // ncols
    colid = i % ncols
    # write row/col indices as axes' title for identification
    axi.set_title("Row:"+str(rowid)+", Col:"+str(colid))
plt.tight_layout()
plt.show()

#plt.imshow(training_images[0])
#print(training_labels[0])
#print(training_images[0])



#normalize to 0~1
training_images  = training_images / 255.0
test_images = test_images / 255.0



#modeling # 3.  
# Flatten: make to 1 dimensional set
# relu: X if X>0 else 0
# softmax: picks the biggest one,
#           for example, [0.1, 0.1, 0.05, 0.1, 9.5, 0.1, 0.05, 0.05, 0.05] ->  [0,0,0,0,1,0,0,0,0]
model = tf.keras.models.Sequential([tf.keras.layers.Flatten(input_shape=(28, 28)),
                                    tf.keras.layers.Dense(128, activation=tf.nn.relu),
                                    tf.keras.layers.Dense(10, activation=tf.nn.softmax)])

model.compile(optimizer = tf.optimizers.Adam(),
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.fit(training_images, training_labels, epochs=5)

#evaluation 6.  
#  (loss)  (accuracy)
loss, accuracy = model.evaluate(test_images, test_labels)
print(" : {}, : {}".format(loss, accuracy))

# 7.  classifications
#         .
#   predictions[0]  9 
#     ankle boot 
predictions = model.predict(test_images)
for i in range(len(training_images)):
    print(" : {}".format(predictions[i]))
    print(" : {}".format(np.argmax(predictions[i])))
```

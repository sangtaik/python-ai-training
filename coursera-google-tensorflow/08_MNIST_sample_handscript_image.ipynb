```python
import tensorflow as tf

# 1. MNIST  
# tensorflow  MNIST     .
# load_data()  x_train, y_train, x_test, y_test   NumPy  .
# x_train, x_test 2828     , 
# y_train, y_test   0~9    .

mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# 2.  
# 0~255.0     0~1.0    .
x_train, x_test = x_train/255.0, x_test/255.0

# 3.  
# tf.keras.models.Sequential()    .

#  (Input layer) Flatten()  2828   784 1  .
#      (Neuron layer) Dense()     (Fully-connected layer) .

#   512 10     
#   (activation function)  ReLU (tf.nn.relu)  (tf.nn.softmax) 
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(512, activation=tf.nn.relu),
    tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])
```

```python
# 4.  
#   .     (Loss function)    optimizer Adam (Adaptive Momentum estimation) .
#   sparse_categorical_crossentropy ,    (accuracy) .
#        .
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# 5.  
# model.fit()   , ,   ,  .
# (epoch) 60,000        .
#     ,     
model.fit(x_train, y_train, epochs=5)

# 6.  
# model.evaluate()  10,000     (loss)  (accuracy) .

#   .
#  : 0.9802
#       98.02%  MNIST       .
# Matplotlib     (accuracy)  (loss)    .
test_loss, test_acc = model.evaluate(x_test, y_test)
print(' :', test_acc)
```

```python
#1  model.evaluate() loss, accuracy  .

loss, accuracy = [], []
for i in range(10):
    model.fit(x_train, y_train, epochs=1)
    loss.append(model.evaluate(x_test, y_test)[0])
    accuracy.append(model.evaluate(x_test, y_test)[1])

print(accuracy)
```

```python
import pandas as pd
```

```python
data1=pd.read_spss("klowf_07_p.sav")
data1.head()
```

## 선택한 변수만 가지고 오기

```python
target1=data1[["Pwave","PwaveB","H05FC","H05LE12","H05LE13"]]
target1.head()
```

## 선택한 변수로 이루어진 데이터프레임의 변수를 한글로 알아보기 쉽게 변경하기

```python
target1.columns=["차수","기존여부","살림살이","1차비용","2차비용"]
target1.head()
```

###  csv  파일 형태로 지정된 폴더에 저장해보기

```python
target1.to_csv("target1.csv",index=False)
```

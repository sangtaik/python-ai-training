```python
# 과제
# 1. 학생들의 이름, 국어, 영어, 수학 성적을 입력 받아 리스트에 저장한 후
# 2. 입력이 종료되면 리스트의 내용을 출력
# 3. 이름을 입력받아 리스트의 내용중 같은 이름의 자료가 존재하면
#    이름, 총점, 평균을 출력하고,
#    없으면 자료 없음이라고 출력
# 4. 입력된 학생들의 인원수가 10명 미만이라면 추가 입력
# 5. 총점의 점수가 가장 작은 학생의 이름, 국어, 영어, 수학 점수 출력
# 6. 전체 학생의 이름, 총점, 평균을 출력 (점수가 높은 순서대로)
# class 변수, class 함수, class Student 사용
class Student:
    
    # 클래스 변수 선언
    count = 0
    students = [] 
    
    #class 함수 선언
    # cls 파라매터는 클래스라는 특수한 고정변수
    @classmethod
    def print(cls):
        print("-----학생 목록-----")
        print("이름\t총점\t평균")
        
        for student in cls.students:
             print(str(student))
        print("------- ------- -------")
        
    @classmethod
    def print_sort_last(cls):
        print("-----가장 총점이 낮은 학생-----")
        print("이름\t총점\t평균")
        
        #총점이 높은대로 정렬
        cls.students.sort()
        
        print(str(cls.students[0]))
        print("------- ------- -------")
        
    @classmethod
    def print_sort_desc(cls):
        print("-----학생 목록-----")
        print("이름\t총점\t평균")
        
        #총점이 높은대로 정렬
        cls.students.sort(reverse=True)
        
        for student in cls.students:
             print(str(student))
        print("------- ------- -------")
    
    def __init__(self, name, korean, math, english, science):
        self.name = name
        self.korean = korean
        self.math = math
        self.english = english
        self.science = science
        Student.students.append(self)
        
        #클래스 변수 초기화
        Student.count += 1
    
    def get_sum(self):
        return self.korean + self.math +\
            self.english + self.science
    def get_average(self):
        return self.get_sum() /4
    
    def to_string(self):
        return "{}\t{}\t{:3.2f}".format(\
            self.name, \
            self.get_sum(),\
            self.get_average())
    ## 특수한 method 선언
    def __str__(self):
        return "{}\t{}\t{:3.2f}".format(\
            self.name, \
            self.get_sum(),\
            self.get_average())
    
    def __lt__(self, value):
        return self.get_sum() < value.get_sum()
    
    @classmethod
    def search_data(cls, name):
        for student in cls.students:
            if student.name == name:
                str(student)
                return student        
        return None
    
    @classmethod
    def input_data(cls):
        while True:
            input_data = input("이름 성적 입력 (종료는 'q'): ").split()
            if input_data[0] == 'q':
                if cls.count < 10:
                    print("{}명 입력이 필요합니다.(최소 10명)".format(10 - cls.count))
                    continue
                else:
                    break
            elif len(input_data) < 4:
                continue
            elif not input_data[1].isdigit():
                continue
            if cls.search_data(input_data[0]) == None:
                print("자료 없음")
                student = Student(input_data[0], int(input_data[1]),
                                  int(input_data[2]), int(input_data[3]), int(input_data[4]))
                for student in cls.students:
                    print(str(student))
            else:
                print("이미 자료가 존재합니다.")
   
Student("윤인성", 87, 98, 88, 95)
Student("연하진", 92, 98, 96, 98)
Student("구지연", 76, 96, 94, 90)
Student("나선주", 98, 92, 96, 92)
Student("윤아린", 95, 98, 98, 98)
Student("윤명월", 64, 88, 92, 92)
Student("서주연", 64, 88, 92, 12)
Student("혼월주", 64, 88, 92, 92)
#Student("김건호", 64, 87, 92, 92)
#Student("성미나", 100, 98, 92, 91)
print()
Student.input_data()
#Student.print()
Student.print_sort_desc()
Student.print_sort_last()
print("현재 생성된 총 학생 수는 {}명입니다.".format(Student.count))
```

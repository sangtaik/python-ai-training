```python
import random
print("# random 모듈")
a_list = [1,2,3,4,5]
print("random : ", random.random()) #( 0 ~ 1사이의 숫장=)
print("uniform : ", random.uniform(10, 20))
print("randrange : ", random.randrange(10)) # 0에서 10 사이의 임의의 수
print("choice : ", random.choice(range(10))) #0에서 10사이
print("shuffle : ", random.shuffle(a_list), a_list) 
print("sample : ", random.sample(a_list, k=2))
```

```python
# 기상청 자료
from urllib import request
from bs4 import BeautifulSoup

target = request.urlopen("http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108")
soup = BeautifulSoup(target, "html.parser")

for location in soup.select("location"):
    print("도시", location.select_one("city").string)
    print("날씨", location.select_one("wf").string)
    print("최저기온", location.select_one("tmn").string)
    print("최고기온", location.select_one("tmx").string)
    print()
#print(soup)
                         
```

```python
# 네이버 영화 평점
from urllib import request
from bs4 import BeautifulSoup

url = "https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date=20220309"
resp = request.urlopen(url)
soup = BeautifulSoup(resp, 'html.parser')
soup


for line in soup.findAll('tr'):
    title = line.find('div', class_='tit5')
    # <td class="title">
    # <div class="tit5">
    # <a href="/movie/bi/mi/basic.naver?code=187320" title="씽2게더">씽2게더</a>
    # </div>
    # </td>
    
    if title :
        movie_title.append(title.get_text().strip('\n'))
        
    # <td class="point">9.38</td>
    point = line.find('td', class_='point')
    if point:
        movie_point.append(point.get_text())

for i, (title, point) in enumerate(zip(movie_title, movie_point)):
    print(i+1 ,  title, ':', point)
        
```

```python
# 날씨
from urllib import request
from bs4 import BeautifulSoup
url = " https://news.naver.com/main/ranking/popularDay.nhn?rankingType=popular_day"
resp = request.urlopen(url)
soup = BeautifulSoup(resp, 'html.parser')

#print(soup)

titles_html = soup.select('.rankingnews_box')

#print(titles_html)

for i in range(len(titles_html)):
    body = titles_html[i].text.split('\n')
    for rows in body:
        print(rows.split(','))
        print()
    
```

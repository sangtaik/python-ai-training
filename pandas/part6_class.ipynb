```python
# merge를 활용하여 대이터프레임 합치기
df1 =pd.read_excel('./dataset/stock price.xlsx') # 주식 가격, 수량
df2 = pd.read_excel('./dataset/stock valuation.xlsx') # 주식비율등.  

print(df1.columns) # <- id, stock_name
print(df2.columns) # <- id, name
```

```python
# 양쪽의 데이터프레임의 'id'컬럼이 같은 자료만 병합, how='inner' 기본설정
pd.merge(df1,df2)  

# how='left' : df1 기준으로 df1은 모두 출력, df2 없으면 Nan
pd.merge(df1,df2, how='left') 

# how='right' : df2 기준으로 df2은 모두 출력, df1 없으면 Nan
pd.merge(df1,df2, how='right') 

# how= 'outer' : df1 ,  df2 모두 출력,  없으면 Nan
pd.merge(df1,df2, how='outer', on='id') # on='key' 양쪽에 있는 컬럼명

# df1의 'name', df2의 'stock_name' 컬럼의 값이 같은 자료를 병합
result = pd.merge(df1, df2, how='left', 
                  right_on='name', left_on='stock_name')
r_data = result.loc[ :, 
                    ['id_x', 'name',  'price','eps']]
r_data
# df1의 price가 50000 미만인 자료와 df2을 결합
pd.merge(df1[df1.price < 50000], df2)
```

```python
# 1. 그룹 만들기
# 2. 그룹별 함수 적용
# 3. 통합한 자료 출력

# 그룹 만들기 : df.groupby(기준이 되는 컬럼명) : 그룹이 여러개면 [ ,.. ]
print("승객수 : ", len(titanic))
# 선실의 레벨(clasee)별로 승객수를 산출 
grouped = titanic.groupby('class')

# 각 그룹별 자료를 출력
for key, group in grouped:
    print(' key : ', key)
    print(' group의 승객 수 : ', len(group))  # 그룹별 승객 수
    print(group.head(3))
    print()
```

```python
# 집계연산을 처리 -> 사용자 함수를 그룹 객체에 적용
df = titanic.loc[ :, ['age','class', 'fare','survived']]
grouped = df.groupby('class')
# group 객체.agg(매핑 함수)
def max_min(x):
    return x.max() - x.min()
# grouped 객체에 max_min 함수를 적용
grouped.agg(max_min)
```

```python
# 모든 열에 여러 함수를 매핑 : 그룹객체.agg( [함수1, 함수2, ...])
# 각 열마다 다른 함수를 매핑 : 그룹객체.agg( {열:함수1, 열1:함수2, ..})
agg_all = grouped.agg( [ 'max', 'min' ])
print(agg_all)

agg_sep = grouped.agg({'age':'mean', 'fare':['min', 'max']})
agg_sep
```

```python
# 그룹객체.transform(매핑함수) : 각 그룹의 행에 대한 연산
age_mean = grouped.age.mean()   # 그룹별 'age'컬럼의 평균
print(age_mean) # 그룹의 key가 인덱스로 설정
age_std = grouped.age.std()   # 그룹별 'age'컬럼의 표준편차
print(age_std)

# 그룹 객체의 age 컬럼에 대해 zscore 계산하여 출력
for key, group in grouped.age:
    group_zscore = (group - age_mean.loc[key]) / age_std.loc[key]
    print(" origin : ", key)
    print(group_zscore.head(3))

def z_score(x):
    return ( x - x.mean())/ x.std()
age_zscore = grouped.age.transform(z_score)
age_zscore.loc[0:9]
```

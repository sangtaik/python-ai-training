```python
import copy

def solution(board, moves):
    answer = 0
    stack = []
    moves_remain = []
    count = 0
    
    moves_remain = copy.deepcopy(moves)
    for i in moves:
        print("\n")
        print(f"--------------------------")

        print(f"---{count}--번째 실행")
        index = i -1
        moves_remain.pop(0)
        print(f"---{moves[count]-1}--열 이동")
        
        for y in range(len(board[index])):
            if (board[y][moves[count]-1] != 0):
                moved_dol = board[y][moves[count]-1]
                print(f"---{moved_dol}--적립")
                stack.append(moved_dol)
                board[y][moves[count]-1] = 0
                break
        
#         for y in range(len(board[index])):
#             index_y = y
#             print(f"index, index_y, board value: {index}, {index_y}, {board[index][index_y]}")
#             if (board[index][index_y] != 0):
#                 moved_dol = board[index][index_y]
#                 print(f"---{moved_dol}--적립")
#                 stack.append(moved_dol)
#                 board[index][index_y] = 0
#                 print(f"---{board[index][index_y]}--0으로 변경")
#                 break
        
        print(f"stack : {stack}")
        if len(stack) > 1:
            prev = stack[len(stack)-2]
            top = stack[-1]
            if prev == top:
                answer += 2
                print(f"crash : {answer}")
                stack.pop()
                stack.pop()
        print(f"남은 stack : {stack}")
        print(f"남은 moves : {moves_remain}")
        print(f"남은 board : {board}")
        count+=1
        print(f"--------------------------")
    print(answer)
    return answer

board = [ [0,0,0,0,0]
         ,[0,0,1,0,3]
         ,[0,2,5,0,1]
         ,[4,2,4,4,2]
         ,[3,5,1,3,1]]
moves = [1,5,3,5,1,2,1,4]

print(solution(board, moves))
```

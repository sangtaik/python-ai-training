경사 하강법(gradient descent)

기울기가 0인 x의 값을 찾아야 하는 것이 목표이다.

어떤 x값으로 구한 기울기와 그 이전에 구한 x값의 기울기를 구하여, 서로 비교하여 기울기가 0인 x의 값을 구한다.

잘못 이동하면 기울기가 오히려 높아진다.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 공부한 시간은 x이며, 2, 4, 6, 8시간 각각 공부했다.
x = np.arange(2, 10, step = 2)
print (x)

# 성적은 81, 93, 91, 97점이 각각 나왔다.
y = np.array([81, 93, 91, 97])
print(y)

# 그래피로 그려본다.
plt.scatter(x, y, s = 100)
plt.show()
```

```python
x_data = np.array(x)
y_data = np.array(y)

# 기울기 a, 절편 b
a = 0
b = 0

# 학습률 learning rate 정하기
lr = 0.03

# 실행횟수 정하기
epochs = 1001

# 경사 하강법
for i in range(epochs):
    y_pred = a * x_data + b # 오차 함수인 y= ax +b 를 정의한 부분
    error = y_data - y_pred # 실체값 - 예측값, 오차를 구하는 식

    # 평균 제곱 오차를 a로 미분한 결과
    a_diff = -(2/len(x_data)) *sum(x_data * (error))

    # 평균 제곱 오차를 b로 미분한 결과
    b_diff = -(2/len(x_data)) * sum(error)

    a = a - lr * a_diff # 미분 결과에 따라서 학습률을 곱한 후 기준의 a값을 업데이트
    b = b - lr * b_diff # 미분 결과에 따라서 학습률을 곱한 후 기준의 b값을 업데이트

    if i % 100 == 0:
        print("epoch=%.f, 기울기=%.04f, 절편=%0.4f" % (i, a, b))

        y_pred = a * x_data + b
        plt.scatter(x, y)
        plt.plot([min(x_data), max(x_data)], [min(y_pred), max(y_pred)])
        plt.show()

print("최종")
print("epoch=%.f, 기울기=%.04f, 절편=%0.4f" % (epoch, a, b))
y_pred = a * x_data + b
plt.scatter(x, y)
plt.plot([min(x_data), max(x_data)], [min(y_pred), max(y_pred)])
plt.show()
```

```python
# 다중 선형 회귀()
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

# 공부한 시간은 x이며, 2, 4, 6, 8시간 각각 공부했다.
x1 = np.arange(2, 10, step = 2)
print (x1)

# 과외 횟수는 0, 4, 2, 3이다.
x2 = np.array([0, 4, 2, 3])
print (x2)

# 성적은 81, 93, 91, 97점이 각각 나왔다.
y = np.array([81, 93, 91, 97])
print(y)

#그래프로 확인해 봅니다.
ax = plt.axes(projection='3d')
ax.set_xlabel('study_hours')
ax.set_ylabel('private_class')
ax.set_zlabel('Score')
ax.dist = 11 
ax.scatter(x1, x2, y)
plt.show()
```

```python
x1_data = np.array(x1)
x2_data = np.array(x2)
y_data = np.array(y)

# 기울기 a1, a2, 절편 b
a1 = 0
a2 = 0
b = 0

# 학습률 learning rate 정하기
lr = 0.02

# 실행횟수 정하기
epochs = 1001

# 경사 하강법
for i in range(epochs):
    y_pred = a1 * x1_data + a2 * x2_data + b
    error = y_data - y_pred
    # 오차 함수를 a1으로 미분한 값
    a1_diff = -(2/len(x1_data)) *sum(x1_data * (error))

    # 오차 함수를 a2으로 미분한 값
    a2_diff = -(2/len(x2_data)) *sum(x2_data * (error))
    # 오차 함수를 B로 미분한 값
    b_diff = -(2/len(x1_data)) * sum(error)

    a1 = a1 - lr * a1_diff
    a2 = a2 - lr * a2_diff
    b = b - lr * b_diff

    if i % 100 == 0:
        print("epoch=%.f, 기울기 a1=%.04f, 기울기 a2=%.04f, 절편=%0.4f" % (i, a1, a2, b))


print("최종")
print("epoch=%.f, 기울기 a1=%.04f, 기울기 a2=%.04f, 절편=%0.4f" % (epochs, a1, a2, b))

```

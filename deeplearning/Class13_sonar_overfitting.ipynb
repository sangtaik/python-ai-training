```python
from keras.backend import binary_crossentropy
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
from keras.models import Sequential
from keras.layers.core import Dense
from sklearn.model_selection import train_test_split

seed = 0
np.random.seed(seed)
tf.random.set_seed(3)

df = pd.read_csv('sonar.csv', header = None)

df.info()

dataset = df.values
X = dataset[:, 0:60].astype(np.float32)
Y_obj = dataset[:, 60]
```

```python

# 문자열에서 숫자로 변경
e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)

print(Y.dtype)
```

```python

# 모델 설정
model = Sequential()
model.add(Dense(24, input_dim=60, activation = 'relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))


model.compile(loss = 'binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.fit(X, Y, epochs=200, batch_size = 5)

print(model.evaluate(X, Y)[1])

# 결과 : 100% 적중률로 과적합 발생!
```

```python
# 과적합 해소 모델, 30%를 테스트 데이터로 남겨둔다.
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3,
                                                    random_state=seed)

# 모델 설정
model = Sequential()
model.add(Dense(24, input_dim=60, activation = 'relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))


model.compile(loss = 'mean_squared_error',
              optimizer='adam',
              metrics=['accuracy'])

model.fit(X_train, Y_train, epochs=130, batch_size = 5)

print(model.evaluate(X_test, Y_test)[1])
```

```python
# 모델 저장
from keras.models import load_model

model.save('my_model.h5')
```

```python
# 저장된 모델 불러오기
model = load_model('my_model.h5')
print(model.evaluate(X_test, Y_test)[1])
```

```python
#
# k겹 교차 검증 실습
#
from sklearn.model_selection import StratifiedKFold


# 학습을 위하여 데이터를 10개로 나눠준다.
n_fold = 10
skf = StratifiedKFold(n_splits = n_fold,
                      shuffle=True, random_state=seed)

accuracy = []

for train, test in skf.split(X, Y):
    
    # 모델 설정
    model = Sequential()
    model.add(Dense(24, input_dim=60, activation = 'relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))


    model.compile(loss = 'binary_crossentropy',
                optimizer='adam',
                metrics=['accuracy'])

    model.fit(X[train], Y[train], epochs=50, batch_size = 5)

    k_accuracy = model.evaluate(X[train], Y[train])[1]
    accuracy.append(k_accuracy)

print(n_fold, accuracy)
```

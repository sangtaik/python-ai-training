```python
# 최소제곱법
# 선형 회귀에서 필요한 기울기를 구하기 위해서 사용한다.
# x값이 1개인 일차 함수일 때 사용이 가능하며, x가 여러 개라면 사용할 수 없다.
# a는 기울기, x는 공부한 시간, y는 출력값 즉 성적이다.
```

```python
# 공부한 시간은 x이며, 2, 4, 6, 8시간 각각 공부했다.
x = np.arange(2, 10, step = 2)
print (x)

# 성적은 81, 93, 91, 97점이 각각 나왔다.
y = np.array([81, 93, 91, 97])
print(y)

# 그래피로 그려본다.
plt.scatter(x, y, s = 100)
plt.show()
```

이미지 1.jpg

```python
# 기울기 a의 공식 (최소 제곱법)
# divisor : 즉 나누는 값, (x- 'x의 평균')**2를 모두 더한 값

# 평균을 구한다.
mx = np.mean(x)
my = np.mean(y)
print(f'x mean = {mx}')
print(f'y mean = {my}')

# 각 (성적 - 평균)의 제곱 값을 구해서 더한다.
dividend = np.sum([(i - mx)*(j-my) for i, j in zip(x,y)])
divisor = np.sum([(i-mx)**2 for i in x])
a = dividend / divisor
b = my - (mx*a)

print('분자 = ', dividend)
print('분모 = ', divisor)
print('기울기 a = ', a)
print('y 절편 b = ', b)
```

```python
# 최소 제곱근으로 직선을 그렸으나, 여러 개의 입력을 처리하기에는 힘들다.
# 따라서 여러 입력에 대하여 임의의 선을 그리고, 이 선이 잘 그어졌는지 평가하여 수정하는 방법을 사용한다.
# 오차를 평가하는 방법인 평균 제곱근을 사용한다.
```

```python

#평균 제곱근

# 공부한 시간은 x이며, 2, 4, 6, 8시간 각각 공부했다.
x = np.arange(2, 10, step = 2)
print (x)

# 성적은 81, 93, 91, 97점이 각각 나왔다.
y = np.array([81, 93, 91, 97])
print(y)

# 그래피로 그려본다.
plt.scatter(x, y, s = 100)
plt.show()
```

```python
# 평균을 구한다.
mx = np.mean(x)
my = np.mean(y)
print(f'x mean = {mx}')
print(f'y mean = {my}')
```

```python
# (x - mx)^2를 이용하여
# Sxx, Sxy를 구한다.

Sxx = np.sum([(i - mx)**2 for i in x])
Sxy = np.sum([(i - mx)*(j-my) for i, j in zip(x,y)])
print(Sxx)
print(Sxy)
```

```python
## 기울기, 절편
slope = Sxy / Sxx
intercept = my - slope * mx

print('기울기 = ', slope)
print('절편 = ', intercept)
```

```python
# 라인을 그려본다.
# y_hat은 y의 예측값
x_ = np.arange(2, 9)
y_hat = intercept + slope*x_
print(x_)
print(y_hat)

plt.scatter(x, y, s=200)
plt.plot(x_, y_hat, c='red')
plt.show()
```

```python
# 평균제곱오차(mean square error) 구하기
mse = 0
for i, j in zip(x, y):
    predicted = intercept + slope * i
    print("공부한 시간 = %.f, 실제 점수= %.f, 예측 점수 = %.f" % (i, j, predicted))
    mse += (j - predicted)**2
print('mse = ', mse/len(x))
```

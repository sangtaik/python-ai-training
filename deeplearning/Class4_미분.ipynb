```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 그래프 공식은 y = x**2로 이미 위에서 정했다. 
x_ = np.arange(-4, 4.2, step=0.1)
y_=x_**2

# x = 4, y = 16부터 시작한다.(탐색한다.)
# 학습률은 0.1이며, 
x = 4
y = 16
lr = 0.1

plt.figure(figsize = (8, 6))
plt.xlabel('x_')
plt.ylabel('y_')

# 선 그래프를 그린다.
plt.plot(x_, y_)

for _ in range(100):

    # 이번 탐색 시작점을 그린다.
    plt.scatter(x, y, c = 'red', s = 100)
    #그래프 공식은 y = x**2로 이미 위에서 정했다. 
    #따라서 도함수 dy_dx는 2*x가 된다.
    dy_dx = 2*x
    # 다음 탐색 위치는 기존 x값에서 lr * dy_dx만큼 -방향으로 움직인다.
    # 이 식을 풀어보면 x1 =  x0 - lr * 2 * x0 이렇게 된다.
    x = x - lr * dy_dx
    # 다음 탐색 위치의 y값은 새로운 x위치의 제곱에 위치한다.
    y = x**2

plt.show()
```
